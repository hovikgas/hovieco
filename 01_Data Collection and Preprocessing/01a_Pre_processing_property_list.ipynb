### Narrowing Down Home Property Examined###

We will use single family residences (comparing a condo or multi unit complex to a single family home does not make sense.

Because of California property tax laws, assessed values are generally based off of purchase price. Homes bought 20 years ago will be assessed at lower value (homes from 1999 in Los Angeles will be generally cheaper than homes from 2019). 

We will pick single family residences from 2010 and later. Home prices will be affected by when it was bought(real estate prices went up from 2010 to 2019). However, we would not have enough homes if we used homes sold in the last 2 years.

Approximately 130,000 single family homes will be included in our assessment.

### Selecting Zipcodes###


- We will be picking Zipcodes with a 100 or more homes.

- We will be using these Zipcodes to evaluate which restaurants to search for based off of zip codes.

- 103 Zipcodes have 100 single family residences or more

- Creating a Zipcode mask (with the 103 Zipcodes) to create a seperate DataFrame


The one common ground we have for the Yelp restaurant list and Property list is Zipcodes. The primary concern is having enough samples for each Zipcodes selected. 

```python
prop_tax_df['ZIPcode5'][(prop_tax_df['LandBaseYear']>2010) & (prop_tax_df['PropertyType']=='SFR')].value_counts()
```

```python
prop_tax_df['ZIPcode5'][(prop_tax_df['LandBaseYear']>2010) & (prop_tax_df['PropertyType']=='SFR')].value_counts()[
    prop_tax_df['ZIPcode5']
    [(prop_tax_df['LandBaseYear']>2010) & (prop_tax_df['PropertyType']=='SFR')].value_counts().values > 100].count()
```

```python
prop_tax_df['ZIPcode5'] = prop_tax_df['ZIPcode5'].astype('int64')
```

```python
prop_tax_df['ZIPcode5'][(prop_tax_df['LandBaseYear']>2010) & (prop_tax_df['PropertyType']=='SFR')].value_counts().index[
    prop_tax_df['ZIPcode5'][(prop_tax_df['LandBaseYear']>2010) & (prop_tax_df['PropertyType']=='SFR')
                        ].value_counts().values > 100].tolist()
```

```python
zip_code_list = prop_tax_df[
    'ZIPcode5'][(prop_tax_df['LandBaseYear']>2010) & (prop_tax_df['PropertyType']=='SFR')
               ].value_counts().index[
    prop_tax_df['ZIPcode5'][(prop_tax_df['LandBaseYear']>2010) & (prop_tax_df['PropertyType']=='SFR')
                           ].value_counts().values > 100].tolist()
```

```python
len(zip_code_list)
```

New Dataframe with only the relevant Zipcodes. 129,000 Single Family Homes still remain.

```python
prop_tax_df[(prop_tax_df['LandBaseYear']>2010)
            & (prop_tax_df['PropertyType']=='SFR') 
            & prop_tax_df['ZIPcode5'].isin(zip_code_list)].shape
```

```python
final_df = prop_tax_df[(prop_tax_df['LandBaseYear']>2010)
            & (prop_tax_df['PropertyType']=='SFR') 
            & prop_tax_df['ZIPcode5'].isin(zip_code_list)]
```
